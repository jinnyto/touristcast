### #3 Check if current data is stationary or not

```python
from statsmodels.tsa.stattools import adfuller

for column in data.columns:
    result = adfuller(data[column])
    print(f'Stationarity test for variable {column}')
    print('ADF Statistic: %f' % result[0])
    print('p-value: %f' % result[1])
    print('Critical Values:')

    for key, value in result[4].items():
        print('\t%s: %.3f' % (key, value))
    
    if result[1] < 0.05:
        print('STATIONARY')
        print(f'{column}')
        print('-------')
    else:
        print('NON-STATIONARY')
        print(f'{column}')
        print('-------')
```

> The initial test shows that only NbDaysOff is a stationary series

### #4 Stationarizing the inputs


#### First trying First Order Difference 
- y_diff = np.diff(y)
- from http://dacatay.com/data-science/part-3-time-series-stationarity-python/

```python
for column in data.columns:
    diff = np.diff(data[column])
    
    diff_result = adfuller(diff)
    print(f'Stationarity test for variable {column}')
    print('ADF Statistic: %f' % diff_result[0])
    print('p-value: %f' % diff_result[1])
    print('Critical Values:')

    for key, value in diff_result[4].items():
        print('\t%s: %.3f' % (key, value))
    
    if diff_result[1] < 0.05:
        print('STATIONARY')
        print(f'{column}')
        print('-------')
    else:
        print('NON-STATIONARY')
        print(f'{column}')
        print('-------')

```

#### Using seasonal_decompose to stationarize data
1. Need to figure out which series' seasonality is additive or multiplicative

```python
from statsmodels.tsa.seasonal import seasonal_decompose

data.index = pd.to_datetime(data.index)

# ----- Additive Model below -----

add_data = seasonal_decompose(data, model='additive')  

add_trend = add_data.trend.dropna()
add_seasonal = add_data.seasonal.dropna()
add_residual = add_data.resid.dropna()

# stationarity_test(add_residual)

# ----- Multiplicative Model below -----

mult_data = seasonal_decompose(data, model='multiplicative')  

mult_trend = mult_data.trend.dropna()
mult_seasonal = mult_data.seasonal.dropna()
mult_residual = mult_data.resid.dropna()

# stationarity_test(mult_residual)
```

2. Comparing to see which series is additive or multiplicative

```python
from statsmodels.tsa.stattools import acf

add_acfs = add_residual.apply(acf, axis=0).apply(np.square, axis=0).apply(np.sum)

mult_acfs = mult_residual.apply(acf, axis=0).apply(np.square, axis=0).apply(np.sum)

additive = add_acfs < mult_acfs
# decomp_type = add_acfs < mult_acfs
# decomp_type[additive] = 'additive'
# decomp_type[~additive] = 'multiplicative'
# decomp_type
add_cols = additive.index[additive].tolist()
mult_cols = additive.index[~additive].tolist()
# fig, axes = plt.subplots(len(data.columns), figsize=(20,50))
# for i, column in enumerate(data.columns):
#     trend[column].plot(ax=axes[i], title=column)
```

#### Combine decomposed data into single dataframe 

```python
resid_data = pd.concat([add_residual[add_cols], mult_residual[mult_cols]], axis=1)
resid_data.info()
```

### #4 Creating VAR model

```python
import statsmodels.api as sm
from statsmodels.tsa.api import VAR, DynamicVAR

model = VAR(resid_data)

results = model.fit(maxlags=12, ic='aic')
results.summary()

# print(model.select_order(12))

# testing data will be NbNights Pays de la Loire from 2016
```
