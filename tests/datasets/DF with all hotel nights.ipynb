### Format total hotel nights dataset downloaded from INSEE
- Remove useless columns and rows
- Rename columns to be easy to read
- Drop rows with 2018 data (incomplete)

**Nb nights**: Number of hotel nights stayed by a tourist in the region

```python
nights = pd.read_csv('/Users/jinny/Documents/touristcast/exercises/datasets/Nuites_hotels_FR_Total_par_region.csv', sep=";")
nights.dropna(inplace=True)
# Remove columns with the word "Codes" in it
nights = nights[nights.columns.drop(list(nights.filter(regex='Codes')))]

# Replace part of column name with NbNights_<RegionName>
nights.columns = [x.strip().replace("Nuitées dans l'hôtellerie - Total - ", 'NbNights_') for x in nights.columns]
nights.columns

# Rename date column and set it to date format
nights = nights.rename(columns={'Libellé': 'Date'})
nights['Date'] = pd.to_datetime(nights.Date)
nights.info()
nights = nights.set_index('Date')

# Remove rows with 2018 data
nights = nights[(nights.index.year != 2018)]
nights
```

```python
nights.to_csv('hotel_nights_FR_by-region.csv')
```
