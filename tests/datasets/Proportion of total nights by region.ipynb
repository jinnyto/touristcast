```python
import pandas as pd
import numpy as np
import datetime 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
nights = pd.read_csv('/Users/jinny/Documents/touristcast/exercises/datasets/hotel_nights_FR_by-region.csv')
nights['Date'] = pd.to_datetime(nights.Date)
nights.info()
nights = nights.set_index('Date')
nights.head()
```

```python
# idf_2010 = sum(nights['NbNights_Île-de-France'][(nights.index.year == 2010)])
# idf_2010

# all_2010 = nights[(nights.index.year == 2010)].sum()

# nightstotal = pd.Series(all_2010)
# nightstotal / (sum(nightstotal))

for n in range(2010,2018):
    allnights = pd.Series(nights[(nights.index.year == n)].sum())
    print(n)
    percent_allnights = allnights / (sum(allnights))
    percent_allnights = percent_allnights.sort_values(ascending=False)
    percent_allnights = percent_allnights.apply(lambda x: f'{x:.2%}')
    print(percent_allnights)
```
