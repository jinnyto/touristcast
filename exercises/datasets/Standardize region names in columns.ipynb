```python
import pandas as pd
import numpy as np
import datetime 
```

```python
stays = pd.read_csv('/Users/jinny/Documents/touristcast/datasets_clean/hotel_nights_FR_by-region_2010-2017.csv')
print(stays.columns)
```

```python
# Replace part of column name with NbNights_<RegionName>
stays.columns = [x.strip().replace("NbNights_", '') for x in stays.columns]
stays.columns
```

```python
stays = stays.rename(columns={'Île-de-France': 'IDF'})
stays = stays.rename(columns={'Centre-Val de Loire': 'CVL'})
stays = stays.rename(columns={'Bourgogne-Franche-Comté': 'BFC'})
stays = stays.rename(columns={'Normandie': 'NOR'})
stays = stays.rename(columns={'Hauts-de-France': 'HDF'})
stays = stays.rename(columns={'Grand Est': 'GES'})
stays = stays.rename(columns={'Pays de la Loire': 'PDL'})
stays = stays.rename(columns={'Bretagne': 'BRE'})
stays = stays.rename(columns={'Nouvelle-Aquitaine': 'NAQ'})
stays = stays.rename(columns={'Occitanie': 'OCC'})
stays = stays.rename(columns={'Auvergne-Rhône-Alpes': 'ARA'})
stays = stays.rename(columns={"Provence-Alpes-Côte d'Azur": 'PAC'})
stays = stays.rename(columns={'Corse': 'COR'})

# stays = stays.add_prefix('nights_')

# stays = stays.rename(columns={'nights_Date': 'date'})

stays.tail()
```

```python
stays.to_csv('nights_2010-2017.csv')
```
