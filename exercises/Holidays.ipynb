```python
# Import workalendar library from https://github.com/peopledoc/workalendar
import datetime as dt
from workalendar.europe import France
```

```python
# Getting a list of holidays for a year with Workalendar
cal = France()
cal.holidays(2012)
```

```python
# Getting the number of days off in a month in a specific year, in this example June 2016

cal.get_working_days_delta(dt.date(2016, 6, 1), dt.date(2016, 6, 30))
from calendar import monthrange

holidays_6_2016 = monthrange(2016, 6)[1] - cal.get_working_days_delta(dt.date(2016, 6, 1), 
                                                                      dt.date(2016, 6, 30))
print(holidays_6_2016)
```

```python
# Getting number of days off in France per month for specific year(s)

def daysoff(y,m):
    nbdays = monthrange(y, m)[1]
    return(nbdays - cal.get_working_days_delta(dt.date(y, m, 1), dt.date(y, m, nbdays)))

# daysoff(2018,5)

# for y in range(2016,2019):
#     for m in range(1,13):
#         print(daysoff(y,m))
    
# list_daysoff = [daysoff(y,m) for y in range(2016,2019) for m in range(1,13)]
# list_daysoff


# Manually change range of years you want to calculate for on the following line:
dict_daysoff = {dt.date(y, m, 1): daysoff(y,m) for y in range(2016,2019) for m in range(1,13)}
dict_daysoff

import pandas as pd

daysoff_series = pd.Series(dict_daysoff)
daysoff_series
```

```python
daysoff_df = daysoff_series.to_frame()
daysoff_df.columns = ['NbDaysOff']
daysoff_df
```
