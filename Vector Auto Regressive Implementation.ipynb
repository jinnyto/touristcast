# VAR Implementation 

```python
##Import Libraries
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
%matplotlib inline
from datetime import datetime as dt
from functools import reduce
```

```python
#Import raw datasets : Night stays between 2010 and 2017
stays_regions = pd.read_csv("touristcast_global_DS.csv", sep=",", index_col = 0)
```

```python
##Data Processing : 
#1 - Select data from 2016 to 2017
stays_regions = stays_regions.loc['2010-01':'2017-12']

#2 - Split data by regions 
##We are not using columns containg Camping stays from raw initial data 'cause some values are missing
##We are also not using Heritage and Aeroports number because the values are constant over time and useless for 
##VAR and SARIMAX coefficient optimization

data_IDF = stays_regions[[ 'FR-IDF_NH_ET', 'FR-IDF_NH_FR', 'FR-IDF_AvgTemp', 'NbDaysOff']]
data_ARA = stays_regions[[ 'FR-ARA_NH_ET', 'FR-ARA_NH_FR', 'FR-ARA_AvgTemp', 'NbDaysOff']]
data_GES= stays_regions[[ 'FR-GES_NH_ET', 'FR-GES_NH_FR', 'FR-GES_AvgTemp', 'NbDaysOff']]
data_BRE= stays_regions[['FR-BRE_NH_ET', 'FR-BRE_NH_FR', 'FR-BRE_AvgTemp', 'NbDaysOff']]
data_CVL= stays_regions[[ 'FR-CVL_NH_ET', 'FR-CVL_NH_FR', 'FR-CVL_AvgTemp', 'NbDaysOff']]
data_COR= stays_regions[['FR-COR_NH_ET', 'FR-COR_NH_FR', 'FR-COR_AvgTemp', 'NbDaysOff']]
data_GES= stays_regions[[ 'FR-GES_NH_ET', 'FR-GES_NH_FR', 'FR-GES_AvgTemp', 'NbDaysOff']]
data_GES= stays_regions[[ 'FR-GES_NH_ET', 'FR-GES_NH_FR', 'FR-GES_AvgTemp', 'NbDaysOff']]
data_NOR= stays_regions[[ 'FR-NOR_NH_ET', 'FR-NOR_NH_FR', 'FR-NOR_AvgTemp', 'NbDaysOff']]
data_NAQ= stays_regions[[ 'FR-NAQ_NH_ET', 'FR-NAQ_NH_FR', 'FR-NAQ_AvgTemp', 'NbDaysOff']]
data_OCC= stays_regions[[ 'FR-OCC_NH_ET', 'FR-OCC_NH_FR', 'FR-OCC_AvgTemp', 'NbDaysOff']]
data_PDL= stays_regions[[ 'FR-PDL_NH_ET', 'FR-PDL_NH_FR', 'FR-PDL_AvgTemp', 'NbDaysOff']]
data_PAC= stays_regions[[ 'FR-PAC_NH_ET', 'FR-PAC_NH_FR', 'FR-PAC_AvgTemp', 'NbDaysOff']]
```

VAR Modelization by Regions

```python
##We are aiming to store predicted values of stays night for french citizens for each regions 
# 1 - Check Stationarity of timeseries 'FR-IDF_NH_ET', 'FR-IDF_NH_FR', 'FR-IDF_AvgTemp', 'NbDaysOff'

from statsmodels.tsa.stattools import adfuller

def test_stationarity(timeseries):
    
    #Determing rolling statistics
    rolmean = pd.rolling_mean(timeseries, window=8)
    rolstd = pd.rolling_std(timeseries, window=8)

    #Plot rolling statistics:
    orig = plt.plot(timeseries, color='blue',label='Original')
    mean = plt.plot(rolmean, color='red', label='Rolling Mean')
    std = plt.plot(rolstd, color='black', label = 'Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviation')
    plt.show(block=False)
    
    #Perform Dickey-Fuller test:
    print ('Results of Dickey-Fuller Test:')
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    print (dfoutput)

##Dickey Fuller Test Analysis Results : Checking Stationarity of each timeseries variables
test_stationarity(data_IDF['FR-IDF_NH_ET']) #Not Stationary
test_stationarity(data_IDF['FR-IDF_NH_FR'])#Stationary
test_stationarity(data_IDF['FR-IDF_AvgTemp'])#Not Stationaty
test_stationarity(data_IDF['NbDaysOff'])#Stationary

test_stationarity(data_ARA['FR-ARA_NH_ET'])#Stationary
test_stationarity(data_ARA['FR-ARA_NH_FR'])#Stationary
test_stationarity(data_ARA['FR-ARA_AvgTemp'])#Stationary
test_stationarity(data_ARA['NbDaysOff'])#Stationary

test_stationarity(data_BRE['FR-BRE_NH_ET'])#Not Stationary
test_stationarity(data_BRE['FR-BRE_NH_FR'])#Stationary
test_stationarity(data_BRE['FR-BRE_AvgTemp'])#Not stationary
test_stationarity(data_BRE['NbDaysOff'])#Stationary

test_stationarity(data_CVL['FR-CVL_NH_ET'])#Not Stationary
test_stationarity(data_CVL['FR-CVL_NH_FR'])#Not Stationary
test_stationarity(data_CVL['FR-CVL_AvgTemp'])#Stationary
test_stationarity(data_CVL['NbDaysOff'])#Stationary

test_stationarity(data_COR['FR-COR_NH_ET'])#Not Stationary
test_stationarity(data_COR['FR-COR_NH_FR'])#Not Stationary
test_stationarity(data_COR['FR-COR_AvgTemp'])#Not Stationary
test_stationarity(data_COR['NbDaysOff'])#Stationary

test_stationarity(data_GES['FR-GES_NH_ET'])#Stationary
test_stationarity(data_GES['FR-GES_NH_FR'])#Stationary
test_stationarity(data_GES['FR-GES_AvgTemp'])#Stationary
test_stationarity(data_GES['NbDaysOff'])#Stationary

test_stationarity(data_NOR['FR-NOR_NH_ET'])#Stationary
test_stationarity(data_NOR['FR-NOR_NH_FR'])#Not stationary
test_stationarity(data_NOR['FR-NOR_AvgTemp'])#Not stationary
test_stationarity(data_NOR['NbDaysOff'])#Stationary

test_stationarity(data_NAQ['FR-NAQ_NH_ET'])#Stationary
test_stationarity(data_NAQ['FR-NAQ_NH_FR'])#Not Stationary
test_stationarity(data_NAQ['FR-NAQ_AvgTemp'])#Stationary
test_stationarity(data_NAQ['NbDaysOff'])#Stationary

test_stationarity(data_PDL['FR-PDL_NH_ET'])#Not Stationary
test_stationarity(data_PDL['FR-PDL_NH_FR'])#Not Stationary
test_stationarity(data_PDL['FR-PDL_AvgTemp'])#Not Stationary
test_stationarity(data_PDL['NbDaysOff'])# Stationary

test_stationarity(data_PAC['FR-PAC_NH_ET'])# Stationary
test_stationarity(data_PAC['FR-PAC_NH_FR'])#Stationary
test_stationarity(data_PAC['FR-PAC_AvgTemp'])# Stationary
test_stationarity(data_PAC['NbDaysOff'])# Stationary
```

```python
##Stationarization of data by Differencing
##ts_log_diff = ts_log - ts_log.shift() (Lag -1)


data_IDF['diff-FR-IDF_NH_ET'] = np.log(data_IDF['FR-IDF_NH_ET']) - np.log(data_IDF['FR-IDF_NH_ET']).shift()
data_IDF['diff-FR-IDF_AvgTemp'] = np.log(data_IDF['FR-IDF_AvgTemp'])- np.log(data_IDF['FR-IDF_AvgTemp']).shift()
data_IDF.dropna(inplace=True)

data_BRE['diff-FR-BRE_NH_ET'] = np.log(data_BRE['FR-BRE_NH_ET']) - np.log(data_BRE['FR-BRE_NH_ET']).shift()
data_BRE['diff-FR-BRE_AvgTemp'] = np.log(data_BRE['FR-BRE_AvgTemp']) - np.log(data_BRE['FR-BRE_AvgTemp']).shift()
data_BRE.dropna(inplace=True)

data_CVL['diff-FR-CVL_NH_ET'] = np.log(data_CVL['FR-CVL_NH_ET']) - np.log(data_CVL['FR-CVL_NH_ET']).shift()
data_CVL['diff-FR-CVL_NH_FR'] = np.log(data_CVL['FR-CVL_NH_FR']) - np.log(data_CVL['FR-CVL_NH_FR']).shift()
data_CVL.dropna(inplace=True)

data_COR['diff-FR-COR_NH_ET'] = np.log(data_COR['FR-COR_NH_ET']) - np.log(data_COR['FR-COR_NH_ET']).shift()
data_COR['diff-FR-COR_NH_FR'] = np.log(data_COR['FR-COR_NH_FR']) - np.log(data_COR['FR-COR_NH_FR']).shift()
data_COR['diff-FR-COR_AvgTemp'] = np.log(data_COR['FR-COR_AvgTemp']) - np.log(data_COR['FR-COR_AvgTemp']).shift()
data_COR.dropna(inplace=True)

data_NOR['diff-FR-NOR_AvgTemp'] = np.log(data_NOR['FR-NOR_AvgTemp']) - np.log(data_NOR['FR-NOR_AvgTemp']).shift()
data_NOR.dropna(inplace=True)

data_NAQ['diff-FR-NAQ_NH_FR'] = np.log(data_NAQ['FR-NAQ_NH_FR']) - np.log(data_NAQ['FR-NAQ_NH_FR']).shift()
data_NAQ.dropna(inplace=True)

data_PDL['diff-FR-PDL_NH_ET'] = np.log(data_PDL['FR-PDL_NH_ET']) - np.log(data_PDL['FR-PDL_NH_ET']).shift()
data_PDL['diff-FR-PDL_NH_FR'] = np.log(data_PDL['FR-PDL_NH_FR']) - np.log(data_PDL['FR-PDL_NH_FR']).shift()
data_PDL['diff-FR-PDL_AvgTemp'] = np.log(data_PDL['FR-PDL_AvgTemp']) - np.log(data_PDL['FR-PDL_AvgTemp']).shift()
data_PDL.dropna(inplace=True)

test_stationarity(data_IDF['diff-FR-IDF_NH_ET'])
test_stationarity(data_IDF['diff-FR-IDF_AvgTemp'])
test_stationarity(data_BRE['diff-FR-BRE_NH_ET'])
test_stationarity(data_BRE['diff-FR-BRE_AvgTemp'])
test_stationarity(data_CVL['diff-FR-CVL_NH_ET'])
test_stationarity(data_CVL['diff-FR-CVL_NH_FR'])
test_stationarity(data_COR['diff-FR-COR_NH_ET'])
test_stationarity(data_COR['diff-FR-COR_NH_FR'])
test_stationarity(data_NOR['diff-FR-NOR_AvgTemp'])
test_stationarity(data_NAQ['diff-FR-NAQ_NH_FR'])
test_stationarity(data_PDL['diff-FR-PDL_NH_ET'] )
test_stationarity(data_PDL['diff-FR-PDL_NH_FR'] )
test_stationarity(data_PDL['diff-FR-PDL_AvgTemp'])
```
